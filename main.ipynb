```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import nltk
```

```python
data = pd.read_csv('testData2.csv', encoding='cp1252')
data
```

```python
data.info()
```

```python
# cleaning of student ID column

data['Student ID'] = data['Student ID'].map(lambda x: int(str(x).replace("[","").replace("]","")))
```

```python
data
```

```python
# checking if the data has null values or not

data.isnull().values.any()
```

```python
# getting english stopwords

from nltk.corpus import stopwords
english_stopwords = stopwords.words('english')
```

```python
# filtering numerical data
text_data = data.select_dtypes(include=['object'])
```

```python
text_data.head()
```

```python
# removing stopwords from text data

for column in text_data.columns:
    text_data[column] = text_data[column].apply(lambda x: ' '.join([word.lower() for word in x.split() if word.lower() not in (english_stopwords)]))
```

```python
text_data
```

```python
# punctuation removal

import re

def clean_punc(sentence):
    cleaned = re.sub(r'[?|!|\'|"|#|;]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    return cleaned
```

```python
for column in text_data.columns:
    text_data[column] = text_data[column].apply(lambda x: ' '.join([clean_punc(word) for word in x.split() if word.lower()]))
```

```python
text_data.info()
```

```python
text_data
```

```python
def whitespace_normalisation(s):
    result = re.sub(r'\s+', " ", s.strip())
    return result
```

```python
for column in text_data.columns:
    text_data[column] = text_data[column].apply(lambda x: ' '.join([whitespace_normalisation(word) for word in x.split() if word.lower()]))
```

```python
text_data
```

```python
# LEMMATIZATION
```

```python
import nltk
from nltk.stem import WordNetLemmatizer 

# Init the Wordnet Lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
for column in text_data.columns:
    text_data[column] = text_data[column].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split() if word.lower()]))
```
